```python
import matplotlib.pyplot as plt
import numpy as np

# 対象のファイルとラベル設定（順番に合わせて）
file_params = [
    ("1.000000_0.000000.txt", "(1, 0)"),
    ("0.500000_0.000000.txt", "(0.5, 0)"),
    ("0.500000_-1.000000.txt", "(0.5, -1)"),
    ("0.000000_0.000000.txt", "(0, 0)"),
    ("0.000000_-1.000000.txt", "(0, -1)"),
    ("-1.000000_-1.250000.txt", "(-1, -1.25)"),
]

colors = ['blue', 'green', 'red', 'orange', 'purple', 'brown']

plt.figure(figsize=(10, 6))

for (filename, label), color in zip(file_params, colors):
    data = np.loadtxt(filename)
    Sj = data[:, 1]  # 2列目が Sj
    S_notj = data[:, 2]  # 3列目が S_{-j}/(N - 1)

    plt.plot(Sj, S_notj, label=label, color=color)

# 軸ラベル（数式表記）
plt.xlabel(r"Payoff of the focal player, $S_j$", fontsize=23)
plt.ylabel(
    "Average payoff of the co-players,\n" + r"$\sum_{k \ne j} S_k / (N - 1)$",
    fontsize=20
)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)

# 凡例のカスタマイズ
handles, label_texts = plt.gca().get_legend_handles_labels()
handles = [plt.Line2D([0], [0], color='none')] + handles
label_texts = [r"$\chi, \kappa$"] + label_texts

plt.legend(handles, label_texts, bbox_to_anchor=(0.2, 0.9), fontsize=18)

# 保存と表示
plt.savefig("Sj vs Smj_chi_kappa_fig1.eps", format='eps', bbox_inches='tight')
plt.show()
```
