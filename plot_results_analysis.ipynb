```python
import matplotlib.pyplot as plt
from tqdm import tqdm

def plot_qarray_from_file(filename, max_timestep=100000):
    print("Streaming file line-by-line (fast mode)...")

    qarray_data = [[] for _ in range(4)]
    step_count = 0

    with open(filename, "r") as file:
        next(file)  # skip header

        for line in tqdm(file, desc="Reading lines"):
            cols = line.strip().split(",")
            timestep = int(cols[0])
            if timestep > max_timestep:
                break

            for i in range(4):
                qarray_data[i].append(float(cols[4 + i]))
            step_count += 1

    # Plot
    plt.figure(figsize=(12, 6))
    labels = [
        r"$\sum q_{ x^{(\mathrm{C})}, x^{(\mathrm{N})}}/2$",
        r"$\sum q_{ x^{(\mathrm{C})}, x^{(\mathrm{C})}}/2$",
        r"$\sum q_{ x^{(\mathrm{N})}, x^{(\mathrm{N})}}/2$",
        r"$\sum q_{ x^{(\mathrm{N})}, x^{(\mathrm{C})}}/2$"
    ]
    for i in range(4):
        plt.plot(range(step_count), qarray_data[i], label=labels[i])

    plt.xlabel("t", fontsize=26)
    plt.ylabel(r"Probability to take $x^{(\mathrm{C})}$", fontsize=26)
    plt.ylim(0,1)
    plt.xticks(fontsize=24)
    plt.yticks(fontsize=24)

    # 凡例をグラフ右の真ん中より少し下に配置
    plt.legend(
        loc="center left",
        bbox_to_anchor=(0.7, 0.75),
        fontsize=21,
        borderaxespad=0.5,
    )
    plt.tight_layout()
    plt.savefig("Action_Probabilities_normal.eps", bbox_inches="tight", format='eps')
    plt.show()


# 実行
plot_qarray_from_file("simulation_results_normal_only.txt")
```
